```python
# Import the necessary libraries

import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.ticker import StrMethodFormatter, NullFormatter
import seaborn as sns
#Ignore warnings.
import warnings
warnings.filterwarnings('ignore') 
```

```python
# Read the data into a dataframe
data = pd.read_csv("loan.csv", header = 0)
#Determine rows and columns
data.shape
```

```python
## Determing the count of columns which have Null values. 
100*data.isnull().mean()
```

```python
#Removing the columns which have all null values
loan_data=data.dropna(axis=1)
loan_data.shape
```

# Verifying the deletion of Columns, Out of 111 columns, only 43 columns have values

```python
#Determing the row which have 1 or more null values 
loan_data.isnull().sum()
```

```python
#Determing the columns having single value columns.
loan_data.nunique().sort_values()
```

```python
#Removing columns having single value as it will not help in analysis
loan_data.drop(['acc_now_delinq','application_type','policy_code','initial_list_status','delinq_amnt','pymnt_plan'], axis=1, inplace=True)
```

```python
#Removing irrelevant columns
#only keeping columns relevant as to identify these risky loan applicants.
loan_data.drop(["id", "member_id","url","funded_amnt",'out_prncp', 'out_prncp_inv','total_rec_int','total_rec_prncp','total_rec_late_fee','recoveries','collection_recovery_fee'],axis=1,inplace=True)
```

```python
#Removing irrelevant rows
#Concerned with Charged-off and Fully paidloans only.
loan_data = loan_data[loan_data.loan_status != 'Current']
```

```python
loan_data.info()
```

```python
#Determing the strong indicators of default cross refrencing values for Charged off loans
valid_data_charged=valid_data[valid_data['loan_status'] == 'Charged Off'].reset_index()
```

```python
# Remove '%' from 'int_rate' and convert to numeric in one step
valid_data_charged['int_rate'] = pd.to_numeric(valid_data_charged['int_rate'].astype(str).str.strip('%'))
```

```python
# Convert 'issue_d' to datetime and extract year and month in one step
valid_data_charged['issue_d'] = pd.to_datetime(valid_data_charged['issue_d'], format="%b-%y")
valid_data_charged['issue_d_yr'] = valid_data_charged['issue_d'].dt.year
valid_data_charged['issue_d_mth'] = valid_data_charged['issue_d'].dt.strftime('%b')
```

```python


# Group by 'issue_d_yr' and count the number of loans for each year
groupdata = valid_data_charged.groupby('issue_d_yr').size()

# Plot the data
plt.bar(groupdata.index, groupdata.values)
plt.title("Year of Loan Despatch Distribution", fontdict={'fontsize': 20, 'fontweight': 5, 'color': 'Green'})
plt.xlabel("Years", fontdict={'fontsize': 12, 'fontweight': 5, 'color': 'Brown'})
plt.ylabel("Loans despatched", fontdict={'fontsize': 12, 'fontweight': 5, 'color': 'Brown'})
plt.show()
```

```python
# Group by 'issue_d_yr' and count the number of loans for each year
groupdata_month = valid_data_charged.groupby('issue_d_mth').size()

# Plot the data
plt.bar(groupdata_month.index, groupdata_month.values)
plt.title("Month of Loan Despatch Distribution", fontdict={'fontsize': 20, 'fontweight': 5, 'color': 'Green'})
plt.xlabel("Months", fontdict={'fontsize': 12, 'fontweight': 5, 'color': 'Brown'})
plt.ylabel("Loans despatched", fontdict={'fontsize': 12, 'fontweight': 5, 'color': 'Brown'})
plt.show()
```

```python
#Observations
#Maximum defaults have happened in last quarter of 2011 when most loans were issued.
```
